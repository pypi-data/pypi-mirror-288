```python
import numpy as np
import matplotlib.pyplot as plt

from spdm.data import Function
```

```python
T=8.0*np.pi

x0=np.linspace(0,T,32)
y0=np.cos(x0)*np.exp(-x0/T)

y1=Function(x0,y0)

x1=np.linspace(0,T,64)

plt.plot(x0,y0,"ro",label="y0")
plt.plot(x1,y1(x1),"b+--",label="y1")
plt.legend()
```

```python
y3 = Function([0, T], lambda x: np.cos(x)*np.exp(-x/T))

y4 = y3+y1+2.0

plt.plot(x1,y3(x1),"bo-",label=r"$y_3=e^{-x/T}\cos(x)$")
plt.plot(x1,y4(x1),"r-",label=r"$y_4=y_1+y_3+2$")
plt.legend()
```
